# Task 3

```python
rolling_mean_euro = fx['EURUSD'].rolling(window = 7).mean()
rolling_std_euro = fx['EURUSD'].rolling(window = 7).std()
```

```python
rolling_mean_euro[:10]
```

```python
rolling_mean_euro[:10].isna().sum() == 6
```

```python
rolling_std_euro[:10]
```

```python
rolling_std_euro[:10].isna().sum() == 6
```

```python
rolling_mean_pln = fx['USDPLN'].rolling(window = 7).mean()
rolling_std_pln = fx['USDPLN'].rolling(window = 7).std()
```

```python
rolling_mean_pln[:10]
```

```python
rolling_mean_pln[:10].isna().sum() == 6
```

```python
rolling_std_pln[:10]
```

```python
rolling_std_pln[:10].isna().sum() == 6
```

# Task 4

```python
fx['baseline_pred'] = fx['EURUSD'].shift(1)
```

```python
fx
```

```python
absolute_error = (fx['EURUSD'] - fx['baseline_pred']).abs()
```

```python
print(absolute_error.mean())
```

```python
fx['baseline_pred'].notna().sum() == len(fx) - 1
```

# Task 5 

```python
report = {
    'currency pairs':['EURUSD','USDPLN'],
    'rolling window size': 7,
    'mean abs error': absolute_error.mean(),
    'date range': [fx.head(1), fx.tail(1)]}
report
```
